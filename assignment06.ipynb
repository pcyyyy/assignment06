# Import packages: 

```python
import numpy as np
import matplotlib.pyplot as plt
```

# Give data

```python
num = 201
std = 20
a = 2 
b = 10

n = np.random.rand(num)
nn = n - np.mean(n)
x = np.linspace(-100,100,num)
y1 = a * x + nn * std + b 
y2 = a * x + b
```

# Get the sum of x,y1,y2

```python
Sx = sum(x)
Sy1 = sum(y1)
Sxy1 = sum(x*y1)
Sy2 = sum(y2)
Sxy2 = sum(x*y2)
Sxx = sum(x**2)
```

# Compute the parameters  of least square for y1

```python
A1 = (Sxy1-Sx*Sy1/num)/(Sxx-Sx**2/num)
A2 = Sy1/num-A1*Sx/num
print('The value of A1 is %f, A2 is %f'%(A1,A2))
```

# Plot the noisy data (x,y1) 

```python
plt.plot(x, y1, 'b.')
plt.show()
```

# Compute the parameters of least square for y2

```python
B1 = (Sxy2-Sx*Sy2/num)/(Sxx-Sx**2/num)
B2 = Sy2/num-A1*Sx/num
print('The value of A1 is %f, A2 is %f'%(B1,B2))
```

# The least square function

```python
fx = B2+B1*x
```

# Plot the clean data (x,y2) 

```python
plt.plot( x, y2, 'k--')
plt.show()
```

# Plot the line that ﬁts the noisy data by the least square error

```python
plt.plot(x, y2, 'b.',x,fx,'k--')
plt.show()
```
